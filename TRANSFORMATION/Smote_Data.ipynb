```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from imblearn.over_sampling import SMOTE

# 1. Baca data
def load_data():
  df = pd.read_csv('stemming.csv')
  return df
```

```python
df = load_data()
df
```

```python
df = df.drop(columns=['username','Tanggal\nKomentar', 'Jumlah Like', 'Komentar', 'remove_http', 'clean_tweet', 'removing_hastag', 'case_folding', 'stopword_removed', 'tokenized'])
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTE
import pandas as pd

# 1. Ambil kolom teks dan label
X_text = df['stemmed']
y = df['Jenis Sentimen']

# 2. Label Encoding
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# 3. TF-IDF Vectorizer
vectorizer = TfidfVectorizer()
X_vectorized = vectorizer.fit_transform(X_text)

# 4. Bagi data 90:10
X_train, X_test, y_train, y_test, text_train, text_test = train_test_split(
    X_vectorized, y_encoded, X_text, test_size=0.1, stratify=y_encoded, random_state=42
)

# 5. Cek distribusi kelas sebelum SMOTE
print("Distribusi kelas di keseluruhan data:")
print(df['Jenis Sentimen'].value_counts())
print("\nDistribusi kelas (sebelum SMOTE):")
print(pd.Series(le.inverse_transform(y_encoded)).value_counts())

# 6. SMOTE pada data latih
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

# 7. Tampilkan jumlah data per kelas setelah SMOTE
df_resampled = pd.DataFrame(X_resampled.toarray(), columns=vectorizer.get_feature_names_out())
df_resampled['target'] = y_resampled
df_resampled['sentimen'] = le.inverse_transform(y_resampled)

# 8. Cek distribusi kelas setelah SMOTE
print("\nDistribusi kelas (setelah SMOTE):")
print(pd.Series(le.inverse_transform(y_resampled)).value_counts())

print("\nJumlah data latih setelah SMOTE:", X_resampled.shape[0])

# 9. Model Naive Bayes
model = MultinomialNB()
model.fit(X_resampled, y_resampled)

# 10. Prediksi data uji
y_pred = model.predict(X_test)

# 11. Tampilkan confusion matrix dan klasifikasi
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=le.classes_))

# 12. Gabungkan hasil prediksi dengan teks asli (opsional untuk analisis manual)
df_eval = pd.DataFrame({
    'teks': text_test.values,
    'label_asli': le.inverse_transform(y_test),
    'label_prediksi': le.inverse_transform(y_pred)
})

# 13. Tampilkan contoh hasil
print("\nContoh hasil prediksi:")
print(df_eval.head())

# 14. Simpan ke CSV
df_eval.to_csv("hasil_prediksi_dengan_teks.csv", index=False)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import numpy as np

labels = le.classes_
encoded_labels = le.transform(labels)
cm = confusion_matrix(y_test, y_pred, labels=encoded_labels)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)

# Visualisasi Confusion Matrix
plt.figure(figsize=(6, 5))  # Ukuran lebih besar untuk 3x3
disp.plot(cmap='Blues', values_format='d')
plt.title("Confusion Matrix: Positif, Negatif, Netral")
plt.grid(False)
plt.show()

print("\n=== Hasil Data Terbaru ===")
print(df_eval.head())
```
